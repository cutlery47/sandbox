```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Суть эксперимента
### Попытаемся понять, как влияет длина подстроки и длина текста на скорость работы различных алгоритмов поиска подстроки в строке
### В работе протестированы алгоритмы:
### - Naive Search
### - Rabin-Karp Search
### - Boyer-Moore Search
### - Knuth-Morris-Pratt Search
### - Python find()
### Скорость работы алгоритмов проверялась на следующих текстах:
### - Lorem Ipsum (455 симв.)
### - Цепочка ДНК (6165 симв.)
### - Текст "Алисы в стране чудес" (227 055 симв.) 
### Тестирование проводилось на: Ryzen 5 4500U (base: 2.4 GHz, но с бустом), 16GB DDR3 (2667 MHz), Python 3.12


```python
lorem_pat = pd.read_csv('outputs/lorem.csv')
dna_pat = pd.read_csv('outputs/dna.csv')
alice_pat = pd.read_csv('outputs/alice.csv')
lorem_10 = pd.read_csv('outputs/lorem_10.csv')
dna_10 = pd.read_csv('outputs/dna_10.csv')
dna_256 = pd.read_csv('outputs/dna_256.csv')
dna_1000 = pd.read_csv('outputs/dna_1000.csv')
alice_10 = pd.read_csv('outputs/alice_10.csv')
alice_256 = pd.read_csv('outputs/alice_256.csv')
alice_1000 = pd.read_csv('outputs/alice_1000.csv')
```

# Тест №1: Изменение длины подстроки
### В данном тесте длина подстроки для поиска изменятся от 1 до 1000 символов (Где возможно). 
### Для каждой длины подстроки тест прогонялся по 10000 раз для всех текстов.

```python
lorem_pat.plot(
    x='len', 
    y=['naive', 'RK', 'BM', 'KMP', 'python'],
    title='Lorem Ipsum', 
    xlabel='pattern length', 
    ylabel='time (nanoseconds)',
)
```

```python
dna_pat.plot(
    x='len', 
    y=['naive', 'RK', 'BM', 'KMP', 'python'],
    title='DNA Sequence', 
    xlabel='pattern length', 
    ylabel='time (milliseconds)',
)
```

```python
alice_pat.plot(
    x='len', 
    y=['naive', 'RK', 'BM', 'KMP', 'python'],
    title='Alice', 
    xlabel='pattern length', 
    ylabel='time (milliseconds)',
)
```
